* Beğenme (like) ve beğenmeme (dislike) sayılarının ortalamasını bulunuz

* ilk 10 videonun title like ve dislike kolonlarını getiriniz.

```python
df[["title","likes","dislikes"]].head(10)
```

* En çok görüntülenen video hangisidir ?

```python
df[df["views"]==df["views"].max()].iloc[0]
```

* En düşük görüntülenen video hangisidir?

```python
df[df["views"]==df["views"].min()].iloc[0]
```

* En az görüntülenen ilk 10 videonun title izlenme sayısını getir

```python
df.sort_values("views",ascending=True).head(10)
```

* Kategoriye göre beğeni ortalamalarını sıralı şekilde getiriniz.

```python
df.groupby("category_id").agg({"likes":"mean"}).sort_values("likes",ascending=False).head(5)
```

* Kategoriye göre yorum sayılarını yukarıdan aşağıya sıralayınız.

```python
df.groupby("category_id").agg({"comment_count":"sum"}).sort_values("comment_count",ascending=False).head(5)
```

* Her kategoride kaç video vardır ?

```python
df["category_id"].value_counts()
```
